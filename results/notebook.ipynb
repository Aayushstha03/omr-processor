First, we import all the csv files for the results for ER tests.

```python
import pandas as pd

# Specify file paths
file_names = [
    "er_posttest_lab.csv",
    "er_posttest_watrin.csv",
    "er_pretest_lab.csv",
    "er_pretest_watrin.csv"
]

# Load each file into a DataFrame and store them in a dictionary
dataframes = {name.rsplit('.', 1)[0]: pd.read_csv(name) for name in file_names}
print(dataframes)


```

Now, we calculate scores for each student in their pretest and posttest.

```python
for name, df in dataframes.items():
    print(name)
    df['score'] = df.iloc[:, 2:].apply(lambda col: sum(col == True), axis=1)
    print(df[["roll_no","score"]])
```

Now for each student, we check if their performance was better in their posttest.

```python
lab_group_df=pd.merge(dataframes["er_pretest_lab"],dataframes["er_posttest_lab"],on="roll_no",how="inner",suffixes=["_pre","_post"])

# print(lab_group_df[["roll_no","score_pre", "score_post"]])
increase_set=lab_group_df[lab_group_df["score_pre"]<lab_group_df["score_post"]]
decreased_set=lab_group_df[lab_group_df["score_pre"]>lab_group_df["score_post"]]
neutral_set=lab_group_df[lab_group_df["score_pre"]==lab_group_df["score_post"]]


print(increase_set[["roll_no","score_pre", "score_post"]])
print(decreased_set[["roll_no","score_pre", "score_post"]])
print(neutral_set[["roll_no","score_pre", "score_post"]])
```
